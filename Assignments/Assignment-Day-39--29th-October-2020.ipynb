## Basket analysis using Apriori algorithm:

```python
basket = df.groupby(['Transaction','Item'])['Quantities'].sum().unstack().fillna(0)
basket
```

```python
basket_model = basket.applymap(lambda x: 1 if x>0 else 0)
basket_model
```

```python
basket_model.Coffee.sum()
```

## In this case we set 'min_support' to 0.02

```python
popular_sets = apriori(basket_model, min_support=0.02, use_colnames=True)
popular_sets
```

```python
rules = association_rules(popular_sets,metric='lift',min_threshold=1)
rules
```

```python
rules[(rules['lift'] >= 1) & (rules['confidence'] >= 0.5)]
```
