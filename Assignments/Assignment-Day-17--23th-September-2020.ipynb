# Question 1:

Using the given MT cars dataset you need to find outlier with z score.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df = df.set_index('model')
df
```

```python
import seaborn as sns
sns.boxplot(x=df['carb']);
```

```python
from scipy import stats
z = np.abs(stats.zscore(df))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

**Outlier**

```python
print(z[30][10])
```

```python
df.describe()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
(df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))
```

```python
df_1 = df[(z < 3).all(axis=1)] # filter the outliers and get the clean data
```

```python
df_1
```

```python
df.shape
```

```python
df_1.shape
```

```python
df_2 = df_1[~((df_1 < (Q1 - 1.5 * IQR)) |(df_1 > (Q3 + 1.5 * IQR))).any(axis=1)] # filter out the outliers by keeping only valid values
df_2.shape
```
